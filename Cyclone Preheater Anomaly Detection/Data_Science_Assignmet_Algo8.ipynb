```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_excel("/content/sample_data/data (1).xlsx")

# Convert the time column to datetime
data['timestamp'] = pd.to_datetime(data['time'])
data.set_index('timestamp', inplace=True)

# Convert relevant columns to numeric and handle errors by setting them as NaN
columns_to_convert = ['Cyclone_Inlet_Gas_Temp', 'Cyclone_Material_Temp',
                      'Cyclone_Outlet_Gas_draft', 'Cyclone_cone_draft',
                      'Cyclone_Gas_Outlet_Temp', 'Cyclone_Inlet_Draft']
for col in columns_to_convert:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Impute missing values using the mean of each column
imputer = SimpleImputer(strategy='mean')
data[columns_to_convert] = imputer.fit_transform(data[columns_to_convert])

# Select the features for anomaly detection
features = ['Cyclone_Inlet_Gas_Temp', 'Cyclone_Material_Temp',
            'Cyclone_Outlet_Gas_draft', 'Cyclone_cone_draft',
            'Cyclone_Gas_Outlet_Temp', 'Cyclone_Inlet_Draft']

# Apply Isolation Forest for anomaly detection
model = IsolationForest(contamination=0.01, random_state=42)
data['anomaly'] = model.fit_predict(data[features])

# Anomalies are labeled as -1, normal points as 1
anomalies = data[data['anomaly'] == -1]

# Plot anomalies for one of the variables
plt.figure(figsize=(15, 6))
plt.plot(data.index, data['Cyclone_Inlet_Gas_Temp'], label='Inlet Gas Temp')
plt.scatter(anomalies.index, anomalies['Cyclone_Inlet_Gas_Temp'], color='red', label='Anomalies')
plt.title('Cyclone Inlet Gas Temperature with Anomalies')
plt.legend()
plt.show()

# Print the number of anomalies detected
print(f"Number of anomalies detected: {len(anomalies)}")

# Plot the Cyclone Inlet Gas Temperature over time (X-axis: Time, Y-axis: Temperature)
plt.plot(data.index, data['Cyclone_Inlet_Gas_Temp'], label='Cyclone Inlet Gas Temp', color='blue')

# Mark detected anomalies in red
plt.scatter(anomalies.index, anomalies['Cyclone_Inlet_Gas_Temp'], color='red', label='Anomalies', marker='o')

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Cyclone Inlet Gas Temperature (C)')
plt.title('Cyclone Inlet Gas Temperature with Anomalies')

# Display the legend to explain the red markers (anomalies) and the blue line (normal temperatures)
plt.legend()

# Show the plot
plt.show()
# Checking the number of detected anomalies (should be ~1% of the data based on contamination=0.01)
num_anomalies_detected = len(data[data['anomaly'] == -1])
total_data_points = len(data)
anomaly_percentage = (num_anomalies_detected / total_data_points) * 100

print(f"Detected Anomalies: {num_anomalies_detected} out of {total_data_points} data points.")
print(f"Anomaly Percentage: {anomaly_percentage:.2f}% (Expected: 1%)")

# Use decision_function to get anomaly scores
anomaly_scores = model.decision_function(data[features])

# Displaying the scores of the first few data points
print("Anomaly Scores (First 10 data points):")
print(anomaly_scores[:10])

# Convert anomaly_scores to a pandas Series for sorting
anomaly_scores = pd.Series(anomaly_scores)
anomalies_sorted = anomaly_scores.sort_values()

# Display the top 10 most extreme anomalies
print(anomalies_sorted.head(10))
```
